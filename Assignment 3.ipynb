```python
def location(N, row, col):# current location of queen
    return N*(row-1) + col

def can_attack(N, row, col):
    result = []
    # check horizontal
    result += [i for i in range(N*(row-1)+1, N*row+1) if i <= N**2 ]
    # check vertical
    result += [j for j in range(col, N*(N-1)+col+1) if j <= N**2][::N]
    # check diagonal(\)
    coordinate_list = []
    for i in range(1,N+1):
        for j in range(1,N+1):
            if i-j == row-col:
                coordinate_list.append((i,j),)          
    for r,c in coordinate_list:
        if location(N,r,c) <= N**2:
            result.append(location(N,r,c))
    # check diagonal(/)
    coordinate_lst = []
    r = row
    c = col
    while 1 <= r <= N and 1 <= c <= N:
        coordinate_lst.append((r, c),)
        r -= 1
        c += 1
    r = row
    c = col
    while 1 <= r <= N and 1 <= c <= N:
        coordinate_lst.append((r,c),)
        r += 1
        c -= 1
    
    #print(f'coordinate_lst = {coordinate_lst}')
    for r,c in coordinate_lst:
        if location(N,r,c) <= N**2:
            result.append(location(N,r,c))

    return result
    
    

def main():
    N,M = map(int, input().split())
    queen_coordinate = []
    for _ in range(M):
        queen_coordinate += (tuple(map(int, input().split())),)
    attacked_block = []
    for row,col in queen_coordinate:
        attacked_block += can_attack(N, row, col)
        #print(attacked_block)
    #print(set(sorted(attacked_block)))
    print(N**2 - len(set(attacked_block)))
    
main()
```
