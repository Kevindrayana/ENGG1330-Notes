```python
def wiseChoice(hotels):
    result = []
    for i in hotels:
        for j in hotels:
            if i == j:
                continue
            if (i != j and dominate(hotels,j,i)):
                break
        else:
            if i not in result:
                result.append(i)
    return result

def readInput(hotels):
    command = input().split()
    while(command[0] != 'END'):
        name = command[0]
        hotels[name] = []
        for i in command[1:]:
            hotels[name].append(float(i))
        command = input().split()  
    return hotels

def dominate(hotels,A,B): # returns True if A dominates B
    for i in range(len(hotels[A])):
        if hotels[A][i] > hotels[B][i]:
            return False
    for j in range(len(hotels[A])):
        if hotels[A][j] < hotels[B][j]:
            return True
    return False

def main():
    hotels={}
    readInput(hotels)
    print(hotels)
    print(wiseChoice(hotels))
main()
```

```python
def createBoard():
    r, c = 6, 7
    if 'n' == input('Standard game? (y/n): '):
        r, c = int(input('r? (2 - 20): ')), int(input('c? (2 - 20): '))
    return [[''] * c for i in range(r)]

def printBoard(board):
    r, c = len(board), len(board[0])
    spaces = 1
    if r>9 or c>9: spaces = 2 #bigBoard
    x = ''
    for row in range(r-1,-1, -1):
        x += f'{row:>{spaces}}'
        ss = ' '
        if spaces==2: ss = '  '
        for col in range(c):
            x += ss+board[row][col]
        x += ' \n'
    x += ' ' + ' '*spaces
    for col in range(c): x += f'{col:>{spaces}}'+' '
    print(x)

def makeMove(board, player, col):
    r = len(board)
    for row in range(r):
        if board[row][col]!='' and board[row+1][col]=='':
            board[row+1][col] = player
            break
        elif board[row][col]=='':
            board[row][col] = player
            break

            
def check_input(board):
    # if the board is full and the user still input in that board
    r, c = len(board), len(board[0])
    for col in range(c):
        if move == 'e':
            return False
        elif (board[r-1][col] == 'X' or board[r-1][col] == 'O') and int(move) == col and move != 'e':
            return True

def check_win(board, player):
    r, c = len(board), len(board[0])
    # horizontal check
    for row in range(r):
        for col in range(c-3):
            if board[row][col] == board[row][col+1] == board[row][col+2] == board[row][col+3] == player:
                return True
        
    # vertical check
    for row in range(r-3):
        for col in range(c):
            if board[row][col] == board[row+1][col] == board[row+2][col] == board[row+3][col] == player:
                return True
    
    # \ check
    for row in range(r-3):
        for col in range(c-3):
            if board[row+3][col] == board[row+2][col+1] == board[row+1][col+2] == board[row+0][col+3] == player:
                return True
    # / check
    for row in range(r-3):
        for col in range(c-3):
            if board[row][col] == board[row+1][col+1] == board[row+2][col+2] == board[row+3][col+3] == player:
                return True

def check_draw(board):
    r, c = len(board), len(board[0])
    counter = 0
    for row in range(r):
        for col in range(c):
            if board[row][col] == 'X' or board[row][col] == 'O':
                counter += 1
                
    if counter == r*c:
        return True
            
            
board = createBoard()
printBoard(board)

player = 'X'
while True:
    
    while True:
        move = input( 'player'+player+' (col #): ')
        if check_input(board):
            printBoard(board)
            pass
        else:
            break
              
    if move == 'e': break
    makeMove(board, player, int(move))
    printBoard(board)
    
    if check_draw(board):
        print("Draw!")
        break
    
    if check_win(board, player):
        print(f"Player {player} has won!")
        break
    if player == 'X': player = 'O'
    else: player = 'X'
print('bye')
```

```python
def wrap(string, max_width):
    result = ''
    counter = 0
    for i in string:
        print(i, end='')
        counter += 1
        if counter % max_width == 0:
            print('\n')
            
    return ''

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```
